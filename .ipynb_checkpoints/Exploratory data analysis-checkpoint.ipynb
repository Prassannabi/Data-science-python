```python
import pandas as pd
```

```python
credit = pd.read_csv('C:\Python assignment\default of credit card clients.csv',header = 1)
```

Analysis of Dataset:

There are 25 variables:

ID: ID of each client

LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit

SEX: Gender (1=male, 2=female)

EDUCATION: (1=graduate school, 2=university, 3=high school, 0,4,5,6=others)

MARRIAGE: Marital status (1=married, 2=single, 3=divorce,0=others)

AGE: Age in years

PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, ... 8=payment delay for eight months, 9=payment delay for nine months and above)

PAY_2: Repayment status in August, 2005 (scale same as above)

PAY_3: Repayment status in July, 2005 (scale same as above)

PAY_4: Repayment status in June, 2005 (scale same as above)

PAY_5: Repayment status in May, 2005 (scale same as above)

PAY_6: Repayment status in April, 2005 (scale same as above)

BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)

BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)

BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)

BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)

BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)

BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)

PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)

PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)

PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)

PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)

PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)

PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)

default.payment.next.month: Default payment (1=yes, 0=no)

##### Renaming column names
The BILL_AMT & PAY_AMT range from 1 to 6 for the six given months. However, a different column PAY ranges starts from 0 and then continues from 2 to 6. To introduce uniformity in the data, we will rename PAY_0 to PAY_1.

The column 'default.payment.next.month' is a little too long. Changing it to Is_Default

```python
credit = credit.rename(columns={'default payment next month' : 'Is_Default', 'PAY_0': 'PAY_1'})
```

```python
from matplotlib import pyplot as plt
```

```python
## renaming the numbers to relevant status
credit['Is_Default'].replace(0, 'No Default' , inplace = True)
credit['Is_Default'].replace(1, 'Default' , inplace = True)
```

```python
## Educationn distribution of defaulters and non defaulters
plt.title('Educationn Distribution')
echart = sns.countplot(x= 'EDUCATION', hue = 'Is_Default', data = credit)
echart.set_xticklabels(['Others','Graduate School','University','High School'])
plt.show()
```

```python
## Gender distribution of defaulters and non defaulters
plt.title('Gender Distribution')
gchart = sns.countplot(x= 'SEX', hue = 'Is_Default', data = credit)
echart.set_xticklabels(['Male','Female'])
plt.show()
```

There isn't a major difference in the distribution of defaulters corresponding to gender, though female has more probability of default than male.

```python
## Marriage distribution of defaulters and non defaulters
plt.title('Marriage Distribution')
gchart = sns.countplot(x= 'MARRIAGE', hue = 'Is_Default', data = credit)
echart.set_xticklabels(['Others','Married','Single','Divorce'])
plt.show()
```

```python
## Age Distribution
sns.set(rc={'figure.figsize':(12,5)})
plt.title('Age Distribution')
gchart = sns.countplot(x= 'AGE', hue = 'Is_Default', data = credit)
plt.show()
```

As the age increases to 30, the probability of default increases. However, when cliets are over 30 the probability decreases.

```python
## Analysing the credit amount
credit['LIMIT_BAL'].describe()
```

As the max value of LIMIT_BAL is 10,00,000 we will group this into 5 bins.

```python
credit['Limitamt_bins'] = pd.cut(x=credit['LIMIT_BAL'], bins=[0,200000,400000,600000,800000,1000000])
```

```python
sns.set(rc={'figure.figsize':(12,5)})
plt.title('Credit Amount Distribution')
gchart = sns.countplot(x= 'Limitamt_bins', hue = 'Is_Default', data = credit)
plt.show()
```

From the above chart, it is evident that clients with lower amount tend to default
